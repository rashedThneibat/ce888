```python
user_features_df = pd.read_excel('movies_latent_factors.xlsx', sheet_name=1)
user_features_df
```

```python
latent_user_preferences = user_features_df.drop(['User'],axis = 1)
latent_user_preferences.shape #,'user_id','key'
```

```python
latent_item_features = item_features_df.drop(['Movie ID', 'Title'],axis = 1)
latent_item_features.shape   # ,'item_id','key'
```

```python
predictions = latent_user_preferences.dot(latent_item_features.T)
```

```python
predictions.columns= item_features_df['Movie ID'].values
```

```python
predictions.index = user_features_df.User.values
```
